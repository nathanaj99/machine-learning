### ASIDE: Functions used for Binary Classification

1. **Split_timeseries**:
    - INPUT: n number of splits
    - OUTPUT: a training set and test set with n number of splits. This means that the number of instances stays the same, but the features multiply by n

2. **important_features(df, multi)**:
    - INPUT: dataframe to extract the 3 important features (max, mean, std) as well as if it is used to multiclass regression or binary
    - OUTPUT: dataframe with extracted features

3. **split_Xy(df, scale, normalize) (ONLY BINARY)**:
    - INPUT: dataframe to be split, including if you want to scale or normalize X
    - OUTPUT: a split X and y from the dataframe, X having the 3 important features and y being the 'bending' column
    
4. **metrics_display(model, X, y) (ONLY BINARY)**:
    - INPUT: model that already learned from training data, X and y for the model to predict
    - OUTPUT: Print functions for the model's accuracy, confusion matrix, ROC curve, and AUC

```python
def split_timeseries(n):
    path = '../data/'

    train_df = pd.DataFrame([])
    test_df = pd.DataFrame([])

    for subdir, dirs, files in os.walk(path):
        action = subdir.split("/")[-1]
        if 'bending' in action:
            test = ['dataset1.csv', 'dataset2.csv']
        else:
            test = ['dataset1.csv', 'dataset2.csv', 'dataset3.csv']
        for file in files:
            path = str(os.path.join(subdir, file))
            if 'csv' in path:
                df = pd.read_csv(path, skiprows=[0, 1, 2, 3])
                
                # SPLIT DATAFRAME
                inc = int(np.round(len(df)/ float(n)))
                start = 0
                end = inc
                
                combined = pd.DataFrame([])
                for i in range(int(n)):
                    if i < n-1:
                        splice = df.loc[start:end]
                    else:
                        splice = df.loc[start:]
                    start = end
                    end += inc
                    
                    splice = splice.describe().drop(columns=['# Columns: time']).transpose().drop(columns=['count']).rename(columns = {'75%': '3rdQ', '25%': '1stQ', '50%': 'median'}).reset_index().drop(columns=['index'])
                    splice = splice.stack().to_frame().T
                    splice.columns = ['{type}{index}'.format(index=c[0]+1, type=c[1]) for c in splice.columns]
                    combined = combined.append(splice)
                combined = combined.reset_index().drop(columns=['index']).stack().to_frame().T
                combined['action'] = action
                if any(num in file for num in test):
                    test_df = test_df.append(combined)
                else:
                    train_df = train_df.append(combined)
                    #print(train_df)
                    
    #train_df = train_df.reset_index().drop(columns=['index'])
    #test_df = test_df.reset_index().drop(columns=['index'])

    return train_df, test_df
```

```python
def important_features(df, multi=False):
    list_col = [col for col in df.columns if (('1' in col[1] or '2' in col[1] or '6' in col[1]) and ('std' in col[1] or 'max' in col[1] or 'mean' in col[1])) or 'action' in col]
    
    df = df[list_col]
    if not multi:
        df['bending'] = df['action'].map(lambda x: True if 'bending' in x else False)
        df = df.drop(columns=['action'])
    
    return df
```

```python
def split_Xy(df, scale=False, normalize=False):
    X = df.drop(columns=['bending'])
    if scale:
        X = preprocessing.scale(X)
    
    if normalize:
        X = preprocessing.normalize(X)
    
    y = df['bending']
    
    return X, y
```

```python
def metrics_display(model, testX, testy):
    print("Accuracy: " + str(model.score(testX, testy)))

    prediction = model.predict(testX)

    confusion_matrix = skm.confusion_matrix(testy, prediction)
    print("Confusion Matrix:")
    print(confusion_matrix)

    predict_prob = model.predict_proba(testX)
    positive_prob = predict_prob[:, 1]
    fpr, tpr, _ = skm.roc_curve(testy, positive_prob)

    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([-0.02, 1.0])
    plt.ylim([0.0, 1.02])
    plt.title("ROC Curve")
    plt.show()

    auc = skm.auc(fpr, tpr)
    print("Area Under Curve: " + str(auc))
```

### A Note of Caution in Running SMOTE
I have tried running the cell below multiple times, and because of the nature of StratifiedKFold, it would randomize the splits every time. The best results hover at around 7, 17, or 19 splits--usually preferring 7. However, when SMOTE finds 17 and 19 to be the best split, the results for the test data are worse. The results below show the best split at 7. I have purposefully not put in a random_state seed because I wanted to see how the algorithm does randomly. So, if running again, please note these observations if getting different results.

```python
"""
FINDINGS:
1. Normalized dataset, with C=10000 --> 100% on test data!
2. normalized, no C --> Terrible (66%)
3. Not normalized, no C --> Equally as terrible
"""

results = {}
for i in range(1, 21):
    # split dataset to i pieces
    train_df, test_df = split_timeseries(i)

    train_df = important_features(train_df)

    trainX, trainy = split_Xy(train_df, normalize=True)
    #trainX, trainy = split_Xy(train_df)
    
    smote = SMOTE(sampling_strategy=1.0)
    trainX_res, trainy_res = smote.fit_resample(trainX, trainy)
    #trainX_resnorm, trainy_resnorm = smote.fit_resample(trainXnorm, trainynorm)
    
    model = LogisticRegression(C=10000)
    rfe = RFECV(model, cv=StratifiedKFold(5), scoring='accuracy', n_jobs = -1)
    rfe = rfe.fit(trainX_res, trainy_res)
    
    results[i] = (rfe.support_, max(rfe.grid_scores_))
```

## e) Binary Classification Using L1-Penalized Logistic Regression

I investigated two different parameters that changed the results of L1-penalized regression
1. **Class_weight = 'balanced'**-- After using SMOTE in the previous problems, I wanted to see if another form of class balancing might improve the results. I understand this wasn't assigned in the homework, but I wanted to see what would happen anyway
2. **Did not include class_weight**--This is the original intention of problem e)

For both, I cross-validated the inverse regularization parameter C with 100 different options

```python
C_scores = {}
C_bal = {}
C = list(np.logspace(-4, 4, 100))

for split in range(1, 21):
    train_df, test_df = split_timeseries(split)
    train_df = important_features(train_df)

    # standardize trainX
    trainX, trainy = split_Xy(train_df, scale=True)

    logcv = LogisticRegressionCV(Cs=C, cv=StratifiedKFold(5), penalty='l1', n_jobs=-1, solver='liblinear')
    logcv_balance = LogisticRegressionCV(Cs=C, cv=StratifiedKFold(5), penalty='l1', n_jobs=-1, solver='liblinear', class_weight = 'balanced')
    logcv.fit(trainX, trainy)
    logcv_balance.fit(trainX, trainy)
    
    argument = C.index(logcv.C_[0])
    argument1 = C.index(logcv_balance.C_[0])

    C_scores[split] = (logcv.C_[0], np.mean([i[argument] for i in logcv.scores_[True]]))
    C_bal[split] = (logcv_balance.C_[0], np.mean([i[argument1] for i in logcv_balance.scores_[True]]))
    
    
```

### Defining Multi-class specific Functions
1. multiclass_metrics_display(model, X, y):
    - INPUT: same as binary function
    - OUTPUT: Accuracy, confusion matrix, multi-class ROC Curve, and AUC for each class in the legend
    
2. split_Xy_multi(df, scale, normalize):
    - INPUT: same as binary function
    - OUTPUT: X, y split of dataframe, y being the 'action' column instead of 'bending' in the binary case

```python
def multiclass_metrics_display(model, testX, testy):
    accuracy = model.score(testX, testy)
    print("Test Error: " + str(1-accuracy))
    print("Accuracy: " + str(accuracy))

    prediction = model.predict(testX)
    confusion_matrix = skm.confusion_matrix(testy, prediction)
    print("Confusion Matrix: ")
    print(confusion_matrix)


    # --- MULTI-CLASS ROC CURVE ---
    fpr = {}
    tpr = {}
    auc = {}

    predict_prob = model.predict_proba(testX)

    class_list=['bending1', 'bending2', 'cycling', 'lying', 'sitting', 'standing', 'walking']
    testy = label_binarize(testy, classes=class_list)

    for i in range(len(class_list)):
        fpr[i], tpr[i], _ = skm.roc_curve(testy[:, i], predict_prob[:, i])
        auc[i] = skm.auc(fpr[i], tpr[i])


    colors = ['yellow', 'cornflowerblue', 'orange', 'green', 'aqua', 'red', 'black']


    for i in range(len(class_list)):
        plt.plot(fpr[i], tpr[i], color=colors[i],
                 label='ROC curve, class %s (AUC = %0.2f)' % (class_list[i], auc[i]))


    # From my research, I also need to include micro and macro lines
    # SOURCE: https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html
    # --- MICRO ---
    fpr["micro"], tpr["micro"], _ = skm.roc_curve(testy.ravel(), predict_prob.ravel())
    auc["micro"] = skm.auc(fpr["micro"], tpr["micro"])

    # --- MACRO ---
    fpr_agg = np.unique(np.concatenate([fpr[i] for i in range(len(class_list))]))

    # Find average of all ROC curves at this points
    tpr_sum = np.zeros_like(fpr_agg)
    for i in range(len(class_list)):
        tpr_sum += interp(fpr_agg, fpr[i], tpr[i])
    tpr_mean = tpr_sum/len(class_list)

    fpr["macro"] = fpr_agg
    tpr["macro"] = tpr_mean
    auc["macro"] = skm.auc(fpr["macro"], tpr["macro"])

    plt.plot(fpr["micro"], tpr["micro"], label='ROC curve, micro (AUC = %0.2f)' % auc["micro"],
            color='brown', linestyle=':')
    plt.plot(fpr["macro"], tpr["macro"], label='ROC curve, macro (AUC = %0.2f)' % auc["macro"],
            color='deeppink', linestyle=':')


    plt.plot([0, 1], [0, 1], 'k--')

    plt.xlim([-0.02, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylim([0.0, 1.02])
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve for All 7 Classes')
    plt.legend(loc="lower right")
    plt.show()
```

##### 2) Using Multinomial Priors (Need to Cross-Validate for the hyperparameter alpha)
I used GridSearchCV to find the ideal alpha for each splice. The best seems to be 7 splices and alpha = 0.0001

```python
multi_results = {}

for i in range(1, 21):
    train_df, test_df = split_timeseries(i)
    train_df = important_features(train_df, multi=True)
    test_df = important_features(test_df, multi=True)

    trainX, trainy = split_Xy_multi(train_df)
    testX, testy = split_Xy_multi(test_df)

    alphas = {'alpha': np.logspace(-4, 4, 100)}

    multi = MultinomialNB()
    multi_cv = GridSearchCV(estimator=multi, param_grid=alphas, n_jobs=-1, cv=StratifiedKFold(5))
    multi_cv.fit(trainX, trainy)
    
    multi_results[i] = (multi_cv.best_score_, multi_cv.best_params_['alpha'])
```

```python
# Get best score and alpha
best_splice = np.argmax([multi_results[i][0] for i in multi_results]) + 1
best_alpha = multi_results[best_splice][1]
print("Ideal Number of Splices: " + str(best_splice))
print("Best Alpha: " + str(best_alpha))

# Retrain multiNB
train_df, test_df = split_timeseries(best_splice)
train_df = important_features(train_df, multi=True)
test_df = important_features(test_df, multi=True)

trainX, trainy = split_Xy_multi(train_df)
testX, testy = split_Xy_multi(test_df)

multi = MultinomialNB(alpha=best_alpha).fit(trainX, trainy)
multiclass_metrics_display(multi, testX, testy)
```

#### Comparing Gaussian vs. Multinomial Priors
It seems like Gaussian does significantly better than Multinomial, and this could be because the data that we have is continuous. The Gaussian distribution handles continuous variables quite nicely, whereas the Multinomial distribution is better for discrete variables such as number of occurrences.
