```python
from scipy.io.arff import loadarff
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
import sklearn.metrics as sk
import numpy as np
```

```python
# Loading data in

data = loadarff("../Data/column_2C_weka.arff")
df = pd.DataFrame(data[0])
```

## iii. Separating test and training set

```python
"""
Abnormal = 1
Normal = 0
- training set: first 70 rows of class 'normal', and 140 rows of class 'abnormal'
"""

# Convert the dataset to 1 and 0
le = preprocessing.LabelEncoder()
df['class'] = le.fit_transform(df['class'])
# LabelEncoder switched the 1 and 0 assignment, so need to map it back
assignment = {0:1, 1:0}
df['class'] = df['class'].map(assignment)
```

```python
# Separate the test and training set by splitting then joining them later
normal = df[df['class'] == 0]
abnormal = df[df['class'] == 1]

train_set_ab, test_set_ab = abnormal.iloc[:140], abnormal.iloc[140:]
train_set_normal, test_set_normal = normal.iloc[:70], normal.iloc[70:]

train_set = pd.concat([train_set_ab, train_set_normal])
test_set = pd.concat([test_set_ab, test_set_normal])

train_X = train_set.drop(columns='class')
train_y = train_set['class']
test_X = test_set.drop(columns='class')
test_y = test_set['class']
```

# c) Classification using kNN
## i & ii. Vanilla kNN

```python
k = np.linspace(208, 1, num=70)
# Assume k = 3 for now
test_error = []
train_error = []
for i in k:
    neigh = KNeighborsClassifier(n_neighbors = int(i))
    neigh.fit(train_X, train_y)
    test_error.append(1-neigh.score(test_X, test_y))
    train_error.append(1-neigh.score(train_X, train_y))

plt.figure(1)
plt.plot(k, test_error)
plt.title('Test Error vs. K-neighbors')
plt.xlabel('K neighbors')
plt.ylabel('Percentage test error')


plt.figure(2)
plt.plot(k, train_error)
plt.title('Training Error vs. K-neighbors')
plt.xlabel('K neighbors')
plt.ylabel('Percentage training error')
```

```python
# IDEAL k*
smallest = np.argmin(test_error)
print("Ideal k* = " + str(int(k[smallest]))) # Ideal k* = 4
```

```python
# Metrics and Evaluation
neigh = KNeighborsClassifier(n_neighbors = 4)
neigh.fit(train_X, train_y)
prediction = neigh.predict(test_X)

# Confusion Matrix
conf_matrix = sk.confusion_matrix(test_y, prediction)
print("Confusion Matrix:" + str(conf_matrix))

# True positive rate & True Negative Rate
tn, fp, fn, tp = sk.confusion_matrix(test_y, prediction).ravel()
tpr = tp/(tp+fn)
tnr = tn/(tn+fp)
print("True Positive Rate: " + str(tpr))
print("True Positive Rate: " + str(tnr))

# Precision
print("Precision: " + str(sk.precision_score(test_y, prediction)))

# F1 score
print("F1 Score: " + str(sk.f1_score(test_y, prediction)))
```

## iii) Best Test Error Rate (Learning Curve)

```python
def ideal_k(n, train_X, train_y, test_X, test_y):
    k = 1
    min_k = 0
    min_error = 100000
    while k < n:
        neigh = KNeighborsClassifier(n_neighbors = k)
        neigh.fit(train_X, train_y)
        test_error = 1-neigh.score(test_X, test_y)
        if test_error < min_error:
            min_error = test_error
            min_k = k
        k += 5
    return min_error
```

```python
n = np.linspace(10, 210, num=21)
min_errors = []
for i in n:
    # pick floor(n/3) of normal class and n-floor(n/3) of abnormal class
    train_set1 = train_set_normal.iloc[:int(np.floor(i/3))]
    train_set2 = train_set_ab.iloc[:int(i-(np.floor(i/3)))]

    # combine the two together
    buffer = pd.concat([train_set1, train_set2])
    train_X_best = buffer.drop(columns='class')
    train_y_best = buffer['class']

    min_errors.append(ideal_k(int(i), train_X_best, train_y_best, test_X, test_y))
    
plt.plot(n, min_errors)
plt.title("Learning curve: N training samples vs. Best test error rate")
plt.xlabel("N training samples")
plt.ylabel("Best test error rate")
```

### i. B) Changing p in Minkowski Distance (using k* = 6)

```python
from decimal import Decimal

def p_root(value, root): 
    root_value = 1 / float(root) 
    return round(Decimal(value) ** Decimal(root_value), 3) 
  
def minkowski_distance(x, y, **kwargs): 
    return (p_root(sum(np.power(abs(a-b), kwargs["power"]) for a, b in zip(x, y)), kwargs["power"]))

x = np.linspace(0.1, 1, num=10)
f = lambda x: 10**x
p = f(x)
test_error_p = []

for i in p:
    neigh = KNeighborsClassifier(n_neighbors = 6, metric = minkowski_distance, metric_params={"power":i})
    neigh.fit(train_X, train_y)
    test_error_p.append(1-neigh.score(test_X, test_y))

df = pd.DataFrame({'log10(p)': x, 'Test error when k=6': test_error_p})
print(df)

ideal_p = np.argmin(test_error_p)
print(ideal_p)
print("The Best log10(p) is: " + str(x[ideal_p]))

plt.plot(x, test_error_p)
plt.title("log10(p) parameter for Minkowski distance vs. Test error when k=6")
plt.xlabel("log10(p)")
plt.ylabel("Test error when k=6")
```

### i. C) Using Chebyshev Distance

```python
k = np.linspace(1, 196, num=40)
test_error_cheb = []

for i in k:
    neigh = KNeighborsClassifier(n_neighbors = int(i), metric='chebyshev')
    neigh.fit(train_X, train_y)
    test_error_cheb.append(1-neigh.score(test_X, test_y))

df = pd.DataFrame({"k": k, "Test Error": test_error_cheb})    
print(df)

plt.plot(k, test_error_cheb)
plt.xlabel("k Neighbors")
plt.ylabel("Percentage Error Test")
plt.title("Test Error Rate vs. k Neighbors for Chebyshev Distance")

print("Ideal k*: " + str(k[np.argmin(test_error_cheb)]))
```

## ii. Mahalanobis Distance

```python
k = np.linspace(1, 196, num=40)
test_error_mahala = []

for i in k:
    neigh = KNeighborsClassifier(algorithm='brute', n_neighbors = int(i), metric='mahalanobis', metric_params={'V': np.cov(train_X)})
    neigh.fit(train_X, train_y)
    test_error_mahala.append(1-neigh.score(test_X, test_y))

df = pd.DataFrame({"k": k, "Test Error": test_error_mahala})    
print(df)

plt.plot(k, test_error_mahala)
plt.xlabel("k Neighbors")
plt.ylabel("Percentage Error Test")
plt.title("Test Error Rate vs. k Neighbors using Mahalanobis Distance")

print("Ideal k*: " + str(k[np.argmin(test_error_mahala)]))
```

## e) Weighted Decision instead of Majority Decision

```python
k = np.linspace(1, 196, num=40)

# -- Euclidian Distance --
min_error_weight_euc = 100
for i in k:
    neigh = KNeighborsClassifier(n_neighbors = int(i), weights='distance')
    neigh.fit(train_X, train_y)
    buffer = 1-neigh.score(test_X, test_y)
    if buffer < min_error_weight_euc:
        min_error_weight_euc = buffer

print("Best test error")
print("Euclidian distance: " + str(min_error_weight_euc))


# -- Manhattan Distance --
min_error_weight_manh = 100
for i in k:
    neigh = KNeighborsClassifier(n_neighbors = int(i), weights='distance', p=1)
    neigh.fit(train_X, train_y)
    buffer = 1-neigh.score(test_X, test_y)
    if buffer < min_error_weight_manh:
        min_error_weight_manh = buffer

print("Manhattan distance: " + str(min_error_weight_manh))

# -- Chebyshev Distance --
min_error_weight_cheb = 100
for i in k:
    neigh = KNeighborsClassifier(n_neighbors = int(i), weights='distance', metric='chebyshev')
    neigh.fit(train_X, train_y)
    buffer = 1-neigh.score(test_X, test_y)
    if buffer < min_error_weight_cheb:
        min_error_weight_cheb = buffer

print("Chebyshev distance: " + str(min_error_weight_cheb))
```
